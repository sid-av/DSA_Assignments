# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
g = sns.FacetGrid(data, col='Classification')
g = g.map(sns.kdeplot, 'SW')
g = sns.FacetGrid(data, col='Classification')
g = g.map(sns.kdeplot, 'PW')
```

```python
plt.style.use('seaborn-deep')
plt.figure(figsize=(15,15))
sns.scatterplot(data=data,x ='PL', y="PW", hue='Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length and Width of Iris species',fontsize=20,pad=20)
```

```python
data1=data.loc[data['Classification']=='Iris-setosa']
data2=data.loc[data['Classification']=='Iris-versicolor']
data3=data.loc[data['Classification']=='Iris-virginica']
plt.boxplot([data1['SL'],data2['SL'],data3['SL']])
plt.xticks([1, 2, 3], ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.title("Sepal Length Vs Classification ",fontsize=20,pad=20)
plt.xlabel("Classification")
plt.ylabel("Sepal Length")
```
